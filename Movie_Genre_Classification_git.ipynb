```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/train_data.txt',sep=':::',names=['id','label','message'],engine='python')
```

```python
df['message']=df['message'].str.lower()
df
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk import sent_tokenize,word_tokenize
from sklearn.preprocessing import LabelEncoder
```

Removing Punctuation

```python
import string
def remove_punctuation(text):
  punctuations=(string.punctuation)
  return text.translate(str.maketrans('','',punctuations))
```

```python
df['message']=df['message'].apply(remove_punctuation)
```

Stopwords removal

```python
from nltk.corpus import stopwords
def stopwords_remove(text):
  stop_words=set(stopwords.words('english'))
  return ' '.join([char for char in text.split() if char not in stop_words])
```

```python
df['message']=df['message'].apply(stopwords_remove)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.svm import SVC
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
```

```python
pipe=Pipeline([
    ('bow',CountVectorizer(max_features=3000)),
    ('classifer',BernoulliNB())
])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y1_test=train_test_split(df['message'],df['label'])
```

```python
pipe.fit(X_train,y_train)
```

```python
pred=pipe.predict(X_test)
```
